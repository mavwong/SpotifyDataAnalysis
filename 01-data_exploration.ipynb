```python
def column_to_boxplot(input_df:DataFrame, export:bool=False):
    plt.style.use('seaborn')
    
    num_in_df = input_df.select_dtypes(include='number')
    col_names = list(num_in_df.columns)

    plot_per_row = 2
    fig, axes = plt.subplots(
        round(len(col_names)/plot_per_row), 
        plot_per_row, 
        figsize = (15, 30)
    )
    
    for index, name in enumerate(col_names):
        i, j = divmod(index, plot_per_row)
        sns.boxplot(x=num_in_df[name], ax=axes[i, j])

    plt.tight_layout()
    plt.show()
```
